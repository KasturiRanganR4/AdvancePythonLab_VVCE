```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Automobile_data.csv")
# Let's start by exploring the dataset.
#print(df.head())  # To see the first few rows of the dataset

# i) Distribution of two and four-door cars by fuel type
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="num-of-doors", hue="fuel-type")
plt.title("Distribution of Two and Four-Door Cars by Fuel Type")
plt.show()

# ii) Distribution of cars by body style and fuel type
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="body-style", hue="fuel-type")
plt.title("Distribution of Cars by Body Style and Fuel Type")
plt.show()

# iii) Total number of each type of body style cars categorized by fuel type
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="body-style", hue="fuel-type", hue_order=["gas", "diesel"])
plt.title("Total Number of Body Style Cars Categorized by Fuel Type")
plt.show()

# iv) Horsepower by fuel type with respect to drive wheel type
plt.figure(figsize=(10, 26))
sns.boxplot(data=df, x="fuel-type", y="horsepower", hue="drive-wheels")
plt.title("Horsepower by Fuel Type and Drive Wheel Type")
plt.show()
```
